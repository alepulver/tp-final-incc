# Exploratory analysis

In this section we'll explore the sample characteristics, and determine possible features for using in the classifyier later.

```python
import book_classification as bc
import shelve
import pandas
import numpy
import matplotlib.pyplot as plt
```

```python
myShelf = shelve.open("storage_new.db")
aBookCollection = myShelf['aBookCollection']
del myShelf
```

## Feature selection

Let's look at the word distribution across books.

```python
tokenizer = bc.BasicTokenizer()
aPossibleFeatureAnalyzer = bc.PossibleFeatureAnalyzer.from_documents(tokenizer, (b.contents for b in aBookCollection))
```

```python
aDataFrame = aPossibleFeatureAnalyzer.as_dataframe()
print(aDataFrame.describe())
countSeries = aDataFrame['Count']
```

It might be a very skewed distribution with possible outliers. Some numbers and a logarithmic box/density plot:

```python
print(countSeries.skew())
print(countSeries.kurt())
```

```python
plt.figsize(4, 5)
boxplot(countSeries.apply(numpy.log))
```

```python
plt.figsize(6, 4)
countSeries.apply(numpy.log).plot(kind='kde')
```

We can see that the words at the extremes are very rare or stopwords.

```python
df = aDataFrame.sort(columns='Count')
print(df.head(5))
print(df.tail(5))
print(len(df))
```

Now we'll remove the first and last 5% of the data.

```python
anotherPossibleFeatureAnalyzer = aPossibleFeatureAnalyzer.prune_quantiles()
anotherDataFrame = anotherPossibleFeatureAnalyzer.as_dataframe()
anotherCountSeries = anotherDataFrame['Count']
df = anotherDataFrame.sort(columns='Count')
print(df.head(5))
print(df.tail(5))
print(len(df))
```

```python
anotherCountSeries.apply(numpy.log).plot(kind='kde')
```

```python
df['Count'].hist(log=True, bins=100)
```

```python
plt.figsize(4, 5)
boxplot(df['Count'].apply(numpy.log))
```

## Words per author

As we're going to use words for feature selection, let's see how they are distributed through the authors' books.

```python
# TODO: precompute and store all words counts per book
#indexer = aPossibleFeatureAnalyzer.build_indexer()
indexer = anotherPossibleFeatureAnalyzer.build_indexer()
extractor = bc.WordFrequencyExtractor(tokenizer, indexer)
frame_input = []
for book in aBookCollection:
    features = extractor.extract_from(book.contents)
    words = features._token_count
    unique_words = len(features.as_dict())
    frame_input.append([book.title, book.author, words, unique_words])
aBookFrame = pandas.DataFrame(frame_input, columns=["Title", "Author", "Words", "Unique Words"])
```

```python
aBookFrame.describe()
```

```python
plt.figsize(12, 5)
aBookFrame.iloc[:, 2:4].hist(bins=50, color='green')
```

From the following figure, it looks there is a relationship between "Words" and "Unique Words". To be sure we could try linear regression, but that's not the topic of our analysis.

```python
plt.figsize(17, 8)
aBookFrame.groupby('Author').mean().plot(kind='bar', subplots=True)
```

Let's look at how the "Unique Words" distribution varies by "Author".

```python
aBookFrame.boxplot(column=['Unique Words'], by='Author', rot=90)
```
